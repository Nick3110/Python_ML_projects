```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1,2])
y_train=np.array([300,500])
# y_train=np.array([300,500])
print(f"it will show  x&y train {y_train} {x_train}")
```

```python
# m is the number of training examples
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
i=0
x_i=x_train[i]
y_i=y_train[i]
print(f"valiue of x & y train is {x_i},{y_i}")
```

```python
plt.scatter(x_train,y_train,marker='X',c='b')#ploting points using scatter tool
plt.title('housing price')
plt.xlabel('size(1000 sq.ft)')
plt.ylabel('price (1000 Dollar)')
plt.show
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      y (ndarray (m,)): target values
    """
    m = x.shape[0]
    print(m)
    f_wb = np.zeros(m)
    print(f_wb)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w=200, b=100,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
